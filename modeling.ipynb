# Mengimpor pustaka

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.metrics import r2_score
import joblib
from google.colab import files
#from google.colab import drive
#drive.mount('/content/drive')
```

# Membaca dataset

```python

full_df = pd.read_csv('/content/full_df (1).csv')
```

membaca dataset full_df dari file berformat CSV menggunakan library pandas

# Menentukan variabel independen (X) dan dependen (y)

```python
X_full = full_df[["semester","tinggi","berat","imt","usia","latihan_per_minggu","perokok"]]
y_full = full_df["pace"]
```

memisahkan dataset menjadi variabel independen (X) dan variabel dependen (y). Variabel independen X_full terdiri dari fitur semester, tinggi badan, berat badan, indeks massa tubuh (IMT), usia, frekuensi latihan per minggu, dan status perokok. Sementara itu, variabel dependen y_full merepresentasikan nilai pace lari, yang menjadi target prediksi dalam pemodelan regresi.

# Membagi dataset menjadi data latih dan data uji

```python
# split data
X_train, X_test, y_train, y_test = train_test_split(
    X_full,
    y_full,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print("X_test :", X_test.shape)
print("y_test :", y_test.shape)
```

Dataset dibagi menjadi data latih (training) dan data uji (testing) menggunakan fungsi train_test_split. Sebanyak 20% data digunakan sebagai data uji, sedangkan 80% data digunakan sebagai data latih. Parameter random_state = 42 digunakan untuk memastikan proses pembagian data bersifat konsisten dan dapat direproduksi, sehingga hasil pemodelan lebih stabil dan dapat diuji ulang.

# Membangun model Random Forest Regressor

```python

model = RandomForestRegressor(
    n_estimators=200,
    random_state=42
)

model.fit(X_train, y_train)
```

Model Random Forest Regressor dibangun dengan menggunakan 200 pohon keputusan (n_estimators = 200) dan nilai random_state = 42 untuk menjaga konsistensi hasil pelatihan. Selanjutnya, model dilatih (fit) menggunakan data latih X_train dan y_train sehingga mampu mempelajari hubungan antara variabel independen dan variabel dependen dalam memprediksi nilai pace lari.

# Menyimpan model

```python


joblib.dump(model, 'model_trained_fix.pkl')
```

Perintah joblib.dump() digunakan untuk menyimpan model Random Forest Regressor yang telah dilatih ke dalam file berekstensi .pkl. Proses ini memungkinkan model digunakan kembali tanpa perlu melakukan pelatihan ulang, sehingga lebih efisien dalam tahap pengujian maupun implementasi (deployment).
